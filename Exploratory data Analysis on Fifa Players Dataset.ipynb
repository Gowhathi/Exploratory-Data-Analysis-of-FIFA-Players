```python
#Dropping columns unnecessary for analysis
data= df.drop(['ID', 'Photo', 'Flag', 'Club Logo', 'Loaned From', 'Jersey Number', 'Volleys'], axis=1)
data.head()
```

```python
def strip_prefix_suffix(value):
    
    if value==np.nan:
        new_value=float(0)
    else:
        value=str(value).replace('€','')
        if value[-1]=='K':
            new_value=float(value.replace('K',''))
        else:
            new_value=float(value.replace('M',''))*1000
    return new_value
```

```python
data['Value']
```

```python
data['Value']= data['Value'].apply(lambda x : strip_prefix_suffix(x))
```

```python
data['Wage']= data['Wage'].apply(lambda x : strip_prefix_suffix(x))
```

```python
data['Joined']= pd.to_datetime(data['Joined'], format='%b %d, %Y').apply(lambda x: x.year).fillna(0).astype(int)
```

```python
data['Release Clause']= data['Release Clause'].fillna("0").apply(lambda x : strip_prefix_suffix(x))
```

```python
data['Height']= data['Height'].fillna("0'0").apply(lambda x: round(int(str(x).split("'")[1])/12+int(str(x).split("'")[0]),2))
```

```python
data['Weight']= data['Weight'].fillna('0').apply(lambda x: str(x).replace('lbs', '')).astype(float)
```

```python
null_values= data.isnull().sum()
null_values
```

```python
indexes= [i for i in null_values[null_values>0].index]
indexes
```

```python
cols_missing_value=[index for index in indexes if index in data.columns and data[index].dtype!='O']
cols_missing_value
```

```python
for col in cols_missing_value:
    if col!='Contract Valid Until':
        data[col]= data[col].fillna(data[col].mean())
```

## Top 20 players ranked by Overall score and whose contract expires in 2020

```python
contract2020= data[data['Contract Valid Until']=='2020-01-01']
contract2020['Rank']= contract2020['Overall'].rank(method='dense', ascending=False)
```

```python
top20 = contract2020.sort_values(['Rank', 'Value'], ascending=[True, False]).head(20)
top20
```

```python
#average age

print('The average age for this set of players is: ', top20['Age'].mean(), 'years')
```

## OBSERVATIONS:

2.The median Value ranges from €32000K ~ €68000K

Top 5 players by Overall rating for each unique position
